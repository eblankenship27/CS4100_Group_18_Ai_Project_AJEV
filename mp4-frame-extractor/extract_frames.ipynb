```python
import cv2
import os

video_path = "overcooked_recording.mp4"   # input recording
output_folder = "dataset_frames"          # folder for saved images
num_images = 250                          # target number of images


# SETUP

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Determine spacing between frames
frame_interval = total_frames // num_images


# FRAME EXTRACTION

saved_count = 0
frame_index = 0

while cap.isOpened() and saved_count < num_images:

    ret, frame = cap.read()

    if not ret:
        break

    # Save frame at interval
    if frame_index % frame_interval == 0:

        filename = f"{output_folder}/frame_{saved_count:04d}.png"
        cv2.imwrite(filename, frame)

        print("Saved:", filename)

        saved_count += 1

    frame_index += 1


cap.release()
```
